```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('SWaT_Dataset_Attack_v0.xlsx', header=1)
```

```python
print(df.shape)
df.head()
```

```python
# Save 'Normal/Attack' column in a separate variable
normal_attack = df['Normal/Attack']

# Drop ' Timestamp' and 'Normal/Attack' features
df = df.drop([' Timestamp', 'Normal/Attack'], axis=1)

# Preprocessing
from sklearn import preprocessing
processor = preprocessing.MinMaxScaler()
df = processor.fit_transform(df)

# Convert the numpy array back into pandas DataFrame
df = pd.DataFrame(df)

# Concatenate 'Normal/Attack' back to the DataFrame
df = pd.concat([df, normal_attack], axis=1)
```

```python

# Convert the DataFrame to a NumPy array
npy_array = df.values

# Save the NumPy array as a .npy file
np.save('attack_data.npy', npy_array)
```
